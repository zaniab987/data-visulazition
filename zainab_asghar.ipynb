**Checking for Outliers**

```python

Q1 = da['Age'].quantile(0.25)
Q3 = da['Age'].quantile(0.75)
IQR = Q3 - Q1
Inter_QRange=Q3-Q1
print(Q1)
print(Q3)
print(Inter_QRange)
```

```python
lower_bound= Q1-1.5* Inter_QRange
print(lower_bound)
upper_bound= Q3+1.5* Inter_QRange
print(upper_bound)
```

```python
outliers = da[(da['Age'] < lower_bound) | (da['Age'] > upper_bound)]['Age']
print("These values are outliers:")
print(outliers)
print("Number of outliers:", outliers.count())
```

```python

# Plotting the actual data points and
plt.scatter(da['Age'].index, da['Age'] , color='pink', label='age' , s=10)
plt.scatter(outliers.index, outliers, color='blue', label='Outliers',s=10)

plt.title('Outliers in age')
plt.xlabel('Index')
plt.ylabel('Lines of Code')
plt.legend()
plt.show()
```

```python
# in our case we don't have any outliners so there will be no change in our means
# Deleting the outliers
df = da[(da['Age'] >= lower_bound) & (da['Age'] <= upper_bound)]



# Calculating mean after removing outliers
mean_without_outliers = da['Age'].mean()

print("Mean without outliers:", mean_without_outliers)
```
